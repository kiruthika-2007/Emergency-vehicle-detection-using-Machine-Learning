```python
from keras.models import Sequential
from keras.layers import Dense, Conv2D, MaxPooling2D, Flatten
```

```python
model = Sequential()
```

```python
#good CNN architecture for ambulance detection with 2000 images for ambukance and 2000 for not ambulan ce
model.add(Conv2D(32, (3, 3), input_shape=(128, 128, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Conv2D(32, (3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Flatten())
model.add(Dense(units=128, activation='relu'))
model.add(Dense(units=1, activation='sigmoid'))
```

```python
model.build()
model.summary()
```

```python
#data augmentataion
from tensorflow.keras.preprocessing.image import ImageDataGenerator
train_datagen = ImageDataGenerator(rescale=1./255,
                                   shear_range=0.2,
                                   zoom_range=0.2,
                                   rotation_range=20,
                                   width_shift_range=0.2,
                                   height_shift_range=0.2,
                                   fill_mode='nearest',
                                   brightness_range=(0.5, 1.5),
                                   channel_shift_range=0.2,
                                   horizontal_flip=True,
                                   vertical_flip=True
                                   )
test_datagen = ImageDataGenerator(rescale=1./255)
```

```python
train_data = train_datagen.flow_from_directory('/content/drive/MyDrive/ambulance_dataset/train',
                                               target_size=(128, 128),
                                               batch_size=32,
                                               class_mode='binary')
test_data = test_datagen.flow_from_directory('/content/drive/MyDrive/ambulance_dataset/test',
                                             target_size=(128, 128),
                                             batch_size=32,
                                             class_mode='binary')
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
train_data.class_indices
```

```python
model.fit(train_data, epochs=100, validation_data=test_data)
```

```python
model.save('ambulance.h5')
```
